```python
start_time = time.time()
SearchResultsDict={}

for i in range(0,int(pages.text[-2:])):
    linkList=driver.find_elements_by_class_name("jl")
    print('Number of links: ',len(linkList)) 
    pageCurrent=driver.find_element_by_class_name("page.current")#return current page
    pageCurrent=str(pageCurrent.text) #store page numbers in text
    
    pages=driver.find_elements_by_class_name("page")
    pages=str(len(pages))
    
    #give some kind of call response to know where we are on the page
    print('Current Results Page '+pageCurrent+' of '+pages+' pages') 
    for j in range(0,len(linkList)):#Range of links
        try:
            linkList[j].click()#Click on iterator link
            time.sleep(randint(5, 10))     #Random wait time to load
            Posting=driver.find_element_by_class_name("jobDesc") #find where the job posting description lives

            jobPostNum = Posting.get_attribute('id') #return post number id
            jobPostNum=re.findall('\d+', jobPostNum)#return only numbers

            jobPostNum='gla-'+jobPostNum[0]+'-'+str(j)#eliminate potential problems with some double ids
            print('Post '+str(j)+' : '+jobPostNum)
            Description=driver.find_element_by_class_name("jobDesc") #get description
            SearchResultsDict[jobPostNum]=Description.text #Store results in dictionary
            #time.sleep(randint(1, 10))
        except:
            print('post error.')
            continue
        
    with open('SearchResultsDict.json', 'a') as fp: #externally append results in json
        json.dump(SearchResultsDict, fp)
        
    driver.find_element_by_class_name("next").click() #click on the next page
print('Total number of jobs recorded: ',len(SearchResultsDict))
end_time = time.time()
total_time=end_time-start_time
print('Time to execute: ',total_time)
```
